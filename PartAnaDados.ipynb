```python
### Você pode dizer que houve alguma influência do poder aquisitivo na sobrevivência do acidente? 
fig = plt.subplots(figsize=(10,8))

sns.barplot(x='classe_bilhete', y='sobreviveu', data=df)
```

```python
### O protocolo marítimo da época do Titanic tinha foco em dar prioridade a crianças e mulheres,
#   analisando os dados podemos dizer se este protocolo foi seguido? 

# Plotando em um gráfico de colunas para identificar esse fator

fig= plt.subplots(figsize=(10,8))

sns.barplot(x='genero', y='sobreviveu', data=df)
```

```python
age_survived = sns.FacetGrid(df, col='sobreviveu')

age_survived.map(sns.distplot, 'idade')
```

```python
### Ter uma família no barco influenciou negativamente ou positivamente para a sua sobrevivência?

fig, axes = plt.subplots(1,2, figsize=(12,4))

sns.barplot(x='pais_filhos_abordo', y='sobreviveu', data=df, ax=axes[0])
sns.barplot(x='irmaos_conjuges_abordo', y='sobreviveu', data=df, ax=axes[1])
```

```python
### Qual o fator que foi mais importante e menos importante para determinar a sobrevivência 
#   para esta população? 
sns.heatmap(df.corr(), cmap='coolwarm', fmt='.2f', linewidths=0.1,
            vmax=1.0, square=True, linecolor='white', annot=True);
```
